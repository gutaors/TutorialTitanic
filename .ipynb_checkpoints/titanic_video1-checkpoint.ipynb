Para quem tiver curiosidade de saber como gerar uma sub igual à gender_submission, esse é o código.

```python
import pandas as pd
```

```python
# data vai ter arquivo test que é a validação
data = pd.read_csv("test.csv")
```

```python
data.head(30)
```

```python
data['Sex'].value_counts()
```

```python
#este cara aqui pega o campo Sex, onde tem female ele coloca True, como esta astype(int) ele coloca 1
e_feminino = (data['Sex'] == 'female').astype(int)
e_feminino.head()
```

```python
#pega como index o PassengerId lá do data
e_feminino.index = data['PassengerId']
e_feminino.head()
```

```python
e_feminino.name = 'Survived'
e_feminino.head()
```

```python
e_feminino.to_csv('gender_submission.csv', header=True)
```
